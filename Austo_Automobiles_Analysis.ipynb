2.5 Preferred Segment of Customers: In the Data , 3 Segments of cars are given, the the most preferred segment chosen given below.

```python
# prompt: Maximum Price of Every Make vs Total Number of Cars in every Make

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is already loaded as in your provided code
# Group by 'Make' and calculate the maximum price and total count
make_price_count = df.groupby('Make').agg(
    Max_Price=('Price', 'max'),
    Total_Cars=('Make', 'count')
).reset_index()

print(make_price_count)

# Visualize the relationship
plt.figure(figsize=(10, 4))

# Scatter plot of Max Price vs. Total Cars
sns.scatterplot(x='Total_Cars', y='Max_Price', data=make_price_count, s=100)
plt.title('Maximum Price of Every Make vs. Total Number of Cars in Every Make')
plt.xlabel('Total Number of Cars')
plt.ylabel('Maximum Price')
plt.grid(True)

# Add labels for each make (optional, can make the plot cluttered)
for i, row in make_price_count.iterrows():
    plt.text(row['Total_Cars'], row['Max_Price'], row['Make'], fontsize=6)

plt.tight_layout()
plt.show()
```

2.7 The Average Salary among all segments, including the Salaried, Professional and Working Partner determines the Budget of the buying cars. Higher the Total Salary, higher the Budget of the car inspite of Loans etc

```python
# prompt: From above chart, relate Total Salary and Make = Sedan, SUV, Hatchback

# Filter the DataFrame for the specified Makes
makes_of_interest = ['Sedan', 'SUV', 'Hatchback']
df_filtered_make = df[df['Make'].isin(makes_of_interest)]

# Group by 'Make' and calculate the mean of 'Total_salary'
average_salary_by_make = df_filtered_make.groupby('Make')['Total_salary'].mean().reset_index()

# Print the resulting DataFrame
print("Average Total Salary by Make (Sedan, SUV, Hatchback):")
print(average_salary_by_make)

# Visualize the relationship
plt.figure(figsize=(8, 6))
sns.barplot(x='Make', y='Total_salary', data=average_salary_by_make)
plt.title('Average Total Salary for Sedan, SUV, and Hatchback')
plt.xlabel('Car Make')
plt.ylabel('Average Total Salary')
plt.show()
```
